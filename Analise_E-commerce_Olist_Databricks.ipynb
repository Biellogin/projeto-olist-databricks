```python
%sql
-- Criando a nossa tabela de limpeza (Silver)
CREATE OR REPLACE TABLE pedidos_limpos AS
SELECT 
    order_id AS id_pedido,
    customer_id AS id_cliente,
    order_status AS status,
    -- Convertendo os textos originais em formato de data real
    to_timestamp(order_purchase_timestamp) AS data_compra,
    to_timestamp(order_delivered_customer_date) AS data_entrega
FROM pedidos_brutos
WHERE order_status = 'delivered'; -- Filtrando apenas o que foi entregue
```

```python
%sql
-- Analisando quantos pedidos foram feitos por mês
SELECT 
    date_format(data_compra, 'yyyy-MM') AS mes_referencia,
    count(id_pedido) AS total_pedidos
FROM pedidos_limpos
GROUP BY 1
ORDER BY 1;
```

```python
%sql
--- Unindo as tabelas para ver o faturamento mensal (JOIN)
--- Usamos 'p' para pedidos e 'i' para itens
SELECT 
    date_format(p.data_compra, 'yyyy-MM') AS mes,
    round(sum(i.price), 2) AS faturamento_total
FROM pedidos_limpos p
JOIN itens_pedidos i ON p.id_pedido = i.order_id
GROUP BY 1
ORDER BY 1;
```
